# Работа с API

* Изучить документацию API OF ICE AND FIRE

* Получить информацию обо всех книгах и сохранить в pd.DataFrame - 1 балл

* Получить информацию обо всех домах Вестероса (Houses) и сохранить в pd.DataFrame - 1 балл

* Получить информацию обо всех домах Вестероса, у которых есть девиз, и сохранить в pd.DataFrame - 2 балла

*hint: используйте метод requests.get(), используйте в ссылке запроса параметр hasTitles для фильтрации*

```python
import requests
import pandas as pd
```

```python
api_base_url = 'https://anapioficeandfire.com/api/'
```

```python
book_list = requests.get(
    f'{api_base_url}/books'# получаем инфу о книгах
).json()
```

```python
df_books = pd.DataFrame(book_list) # сохраняем в датафрейм
df_books.head()
```

```python
houses_list = requests.get(
    f'{api_base_url}/houses' # получаем инфу о домах
).json()

df_houses = pd.DataFrame(houses_list) # сохраняем в датафрейм
df_houses.head()
```

```python
words_list = requests.get(
    f'{api_base_url}/houses?hasWords=true' # получаем инфу о домах с девизом
).json()

df_words = pd.DataFrame(words_list) # сохраняем в датафрейм
df_words.head()
```
